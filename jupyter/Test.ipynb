```python
import ctypes
import pandas as pd

# Load the shared library
lib = ctypes.CDLL("/usr/local/lib/libFeatureExtractionLib.so")

# Define the argument and return types for the C++ function
my_function = lib.PredictImageExtern
my_function.argtypes = [ctypes.c_char_p]
my_function.restype = ctypes.c_char_p


# Create a Python wrapper function that takes a string argument
def Detect(my_string):
    # Convert the Python string to a C string (encoded in ASCII)
    f = my_string+".wmv"
    c_string = f.encode('ascii')

    # Call the C++ function with the C string argument
    result = my_function(c_string)
    #     # CSV file
    df = pd.read_csv("processed/"+my_string+".csv")
    return df
```
