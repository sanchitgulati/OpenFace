```python
import ctypes
import pandas as pd

# Load the shared library
lib = ctypes.CDLL("/usr/local/lib/libFeatureExtractionLib.so")

# Define the argument and return types for the C++ function
cPredictFile = lib.PredictFile
cPredictFile.argtypes = [ctypes.c_char_p]
cPredictFile.restype = None
cPredictFolder = lib.PredictFolder
cPredictFolder.argtypes = [ctypes.c_char_p]
cPredictFolder.restype = None


def PredictAUs(filename,extention):
    # Convert the Python string to a C string (encoded in ASCII)
    f = filename+"."+extention
    c_string = f.encode('ascii')

    # Call the C++ function with the C string argument
    cPredictFile(c_string)
    #     # CSV file
    df = pd.read_csv("processed/"+filename+".csv")
    return df.filter(like='AU')


def PredictFolder(folder):
    # Convert the Python string to a C string (encoded in ASCII)
    c_string = folder.encode('ascii')

    # Call the C++ function with the C string argument
    cPredictFolder(c_string)
    #     # CSV file
    df = pd.read_csv("processed/"+folder+".csv")
    return df
    
```

```python
filename = "sample1"
extension = "jpg"
df = PredictAUs(filename,extension)
df2 = PredictFolder("test")
```
